```python
import re, random
class Person: 
    def __init__(self, filename): 
        lines = self.readFile(filename)
        self.wordDict, self.startDict = self.generateDict(lines)
        self.alpha = 0.30
        
    def readFile(self, filename): 
        with open(filename) as f:
            lines = f.readlines()
        lines = [re.sub('([.,!?()])', r' \1 ', line.replace("\n", "")).split() for line in lines]
        return lines
    
    def generateDict(self, lines):
        wordDict = {}
        startDict = {}
        for line in lines: 
            for i in range (len(line)):
                word = line[i]
                if i == 0: 
                    if word not in startDict.keys():
                        startDict[word] = 1
                    else:
                        startDict[word] += 1
                    if word not in wordDict.keys():
                        wordDict[word] = {}
                if i > 0:
                    if word not in wordDict.keys():
                        wordDict[word] = {}
                    if word in wordDict[line[i-1]].keys():
                        wordDict[line[i-1]][word] += 1
                    else:
                        wordDict[line[i-1]][word] = 1
        
        # normalizing the wordDict
        for word in wordDict:
            aveLen = sum(wordDict[word].values())
            for key in wordDict[word].keys(): 
                wordDict[word][key] /= aveLen

        # normalize startDict
        aveLen = sum(startDict.values())
        for key in startDict.keys():
            startDict[key] /= aveLen
    
        return wordDict, startDict
    
    def generateSentence(self, context = None): 
        lastWord = ""
        if context == None: 
            lastWord = self.generateStartWord()
            output = ""+lastWord

            while(True):
                prob = random.random()
                acc = 0
                dictionary = self.wordDict[lastWord]
                for word in dictionary.keys(): 
                    acc += dictionary[word]
                    if prob < acc: 
                        lastWord = word
                        if lastWord not in [',', '.', '!', '?']: 
                            output += " " + word
                        else: 
                            output += word
                        break
                if lastWord in ['.', '!', '?']:
                    break

            return output
        
        else: 
            frequencyDict = {}
            for word in context: 
                if word in self.wordDict.keys(): 
                    for key in self.wordDict[word].keys(): 
                        if key in frequencyDict.keys():
                            frequencyDict[key] += self.wordDict[word][key]
                        else: 
                            frequencyDict[key] = self.wordDict[word][key]
            words = list({k: v for k, v in sorted(frequencyDict.items(), key=lambda item: item[1])}.keys())
            if words == [] or random.random() < self.alpha: 
                return self.generateSentence()
            wordRank2 = words[1]
            if wordRank2 not in self.wordDict.keys(): 
                return self.generateSentence()
            else: 
                while(True): 
                    sentence = self.generateSentence()
                    if wordRank2 in sentence: 
                        return sentence
            

    def generateStartWord(self):
        prob = random.random()
        acc = 0
        for word in self.startDict.keys():
            acc += self.startDict[word]
            if prob < acc: 
                return word
```

```python
Bill = Person("BillSamples.txt")
Bruce = Person("BruceSamples.txt")
```

```python
BillSentence = Bill.generateSentence()
print("Bill:", BillSentence)
BruceSentence = Bruce.generateSentence(BillSentence)
print("Bruce:", BruceSentence)

for i in range(10): 
    BillSentence = Bill.generateSentence(BruceSentence)
    print("Bill:", BillSentence)

    BruceSentence = Bruce.generateSentence(BillSentence)
    print("Bruce:", BruceSentence)
```
